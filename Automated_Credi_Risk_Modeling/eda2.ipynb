```python
import numpy as np
import pandas as pd

filtered_df = df.copy()

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]) and col != 'loan_status':
        Q1 = np.percentile(filtered_df[col], 25, interpolation='midpoint')
        Q3 = np.percentile(filtered_df[col], 75, interpolation='midpoint')
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Calculate median
        median = filtered_df[col].median()

        # Replace outliers with median
        filtered_df[col] = np.where(
            (filtered_df[col] < lower) | (filtered_df[col] > upper),
            median,
            filtered_df[col]
        )
```

```python
Q1, Q3 = np.percentile(df['credit_score'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['credit_score'] = df['credit_score'].clip(*bounds)
```

```python
Q1, Q3 = np.percentile(df['person_age'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['person_age'] = df['person_age'].clip(*bounds)

Q1, Q3 = np.percentile(df['person_income'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['person_income'] = df['person_income'].clip(*bounds)

Q1, Q3 = np.percentile(df['person_emp_exp'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['person_emp_exp'] = df['person_emp_exp'].clip(*bounds)

Q1, Q3 = np.percentile(df['loan_amnt'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['loan_amnt'] = df['loan_amnt'].clip(*bounds)

Q1, Q3 = np.percentile(df['loan_int_rate'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['loan_int_rate'] = df['loan_int_rate'].clip(*bounds)

Q1, Q3 = np.percentile(df['loan_percent_income'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['loan_percent_income'] = df['loan_percent_income'].clip(*bounds)

Q1, Q3 = np.percentile(df['cb_person_cred_hist_length'], [25, 75])
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Step 2: Cap scores to bounds
df['cb_person_cred_hist_length'] = df['cb_person_cred_hist_length'].clip(*bounds)


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in df.columns:
    # Check if column is numerical AND not the target
    if pd.api.types.is_numeric_dtype(df[col]) and col != 'loan_status':
        plt.figure(figsize=(10, 6))
        sns.kdeplot(
            data=df,
            x=col,
            hue='loan_status',
            fill=True,
            palette=['red', 'green'],
            alpha=0.6,  # Transparency for overlapping areas
            common_norm=False  # Normalize each distribution independently
        )
        plt.title(f'Distribution of {col} by Loan Status', pad=20)
        plt.xlabel(col)
        plt.ylabel('Density')
        plt.legend(title='Loan Status', labels=['Rejected (0)', 'Approved (1)'])
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.show()
```

```python
from scipy.stats import mannwhitneyu
group0 = df[df['loan_status'] == 0]['credit_score']
group1 = df[df['loan_status'] == 1]['credit_score']
stat, p = mannwhitneyu(group0, group1)
print(f"p-value = {p:.4f}  {'Significant' if p < 0.05 else 'Not significant'}")

```

```python
group02 = df[df['loan_status'] == 0]['loan_amnt']
group01 = df[df['loan_status'] == 1]['loan_amnt']
stat, p = mannwhitneyu(group02, group01)
print(f"p-value = {p:.4f}  {'Significant' if p < 0.05 else 'Not significant'}")
```

```python
group03 = df[df['loan_status'] == 0]['person_income']
group04 = df[df['loan_status'] == 1]['person_income']
stat, p = mannwhitneyu(group03, group04)
print(f"p-value = {p:.4f}  {'Significant' if p < 0.05 else 'Not significant'}")
```
